```python
# Imports
import os
import pandas as pd
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Fixed files
data_dir = '/content/gdrive/My Drive/NTU/CURRENT SEM/CZ2001 ALGORITHMS/ALGO PROJECT 2'

stats_h = os.path.join(data_dir, '2210_test_benchmark_h.csv')
stats_n = os.path.join(data_dir, '2210_test_benchmark_n.csv')
stats_k = os.path.join(data_dir, '3110_test_benchmark_k.csv')
num_hos = os.path.join(data_dir, 'num_of_hospital_vs_time_taken.csv')
```

```python
# Create df
h_df = pd.read_csv(stats_h)
n_df = pd.read_csv(stats_n)
k_df = pd.read_csv(stats_k)
num_hos_df = pd.read_csv(num_hos)
```

# Our test results to match simulation

```python
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(h_df['number_of_hospitals'],  h_df['multi_source_timing'], c='red')
plt.plot(h_df['number_of_hospitals'],  h_df['single_source_timing'], c='blue')
plt.xlabel("Number of Hospitals", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```

```python
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(n_df['number_of_nodes'],  n_df['multi_source_timing'], c='red')
plt.plot(n_df['number_of_nodes'],  n_df['single_source_timing'], c='blue')
plt.xlabel("Number of Nodes", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```
