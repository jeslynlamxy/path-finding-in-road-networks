# Our test results to match simulation

```python
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(h_df['number_of_hospitals'],  h_df['multi_source_timing'], c='red')
plt.plot(h_df['number_of_hospitals'],  h_df['single_source_timing'], c='blue')
plt.xlabel("Number of Hospitals", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```

```python
firt_few_h_df = h_df.head(50)
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(firt_few_h_df['number_of_hospitals'],  firt_few_h_df['multi_source_timing'], c='red')
plt.plot(firt_few_h_df['number_of_hospitals'],  firt_few_h_df['single_source_timing'], c='blue')
plt.xlabel("Number of Hospitals", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```

```python
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(n_df['number_of_nodes'],  n_df['multi_source_timing'], c='red')
plt.plot(n_df['number_of_nodes'],  n_df['single_source_timing'], c='blue')
plt.xlabel("Number of Nodes", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```

```python
# sssp vs mssp
plt.figure(figsize=(16, 10), dpi= 80, facecolor='w', edgecolor='k')
plt.plot(k_df['number_of_paths'],  k_df['multi_source_timing'], c='red')
plt.plot(k_df['number_of_paths'],  k_df['single_source_timing'], c='blue')
plt.xlabel("Number of Paths", fontsize=24)
plt.ylabel("Time Taken", fontsize=24)
plt.legend(['Multi Source Single Path', 'Single Source Single Path'], loc='best', fontsize=24)
```
